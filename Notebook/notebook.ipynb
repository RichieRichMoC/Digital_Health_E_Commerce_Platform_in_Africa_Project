# BankMarketing: Predictive Modeling for Term Deposit Subscription in Direct Phone Campaigns

## Introduction

Welcome to the BankMarketing Insights Project! In the highly competitive banking sector, understanding customer behavior is vital for optimizing marketing strategies and driving product adoption. This project focuses on analyzing data from direct marketing campaigns conducted by a banking institution, where clients were contacted via phone calls—often multiple times—to determine their interest in subscribing to a term deposit product. The goal of the BankMarketing Insights Project is to develop a predictive model that can effectively assess the likelihood of a customer subscribing to the term deposit, thereby enhancing the efficiency and targeting of future marketing efforts.

## Business Understanding

For our banking client, increasing the success rate of term deposit subscriptions is a strategic priority in achieving long-term profitability and customer engagement. In an increasingly saturated financial services market, direct marketing—especially via phone campaigns—remains a vital channel to reach potential clients. However, these campaigns can be costly and time-consuming, often requiring multiple contacts with the same customer. Optimizing these efforts by understanding which clients are most likely to subscribe to a term deposit is essential for maximizing marketing ROI and improving operational efficiency.

The primary objective of the BankMarketing Insights Project is to develop a predictive model that accurately forecasts whether a customer will subscribe to a term deposit product, based on historical data collected during past phone-based marketing campaigns. By leveraging machine learning techniques, the model will help the bank identify high-potential leads and tailor outreach strategies accordingly—minimizing wasted effort and enhancing campaign effectiveness.

The dataset includes 16 input variables and 1 output variable. These span demographic, financial, and campaign-related attributes, offering a comprehensive view of each client's profile and their interaction history with the bank's marketing efforts:

Input Variables:
Client Information:

age: Client's age (numeric)

job: Type of job (categorical)

marital: Marital status (categorical)

education: Education level (categorical)

default: Whether the client has credit in default (binary)

balance: Average yearly balance in euros (numeric)

housing: Has a housing loan? (binary)

loan: Has a personal loan? (binary)

Campaign Contact Information:

contact: Contact communication type (categorical)

day: Last contact day of the month (numeric)

month: Last contact month of the year (categorical)

duration: Duration of the last contact in seconds (numeric)

Campaign Performance Metrics:

campaign: Number of contacts during this campaign (numeric)

pdays: Days since last contact from a previous campaign (-1 indicates no previous contact)

previous: Number of contacts before this campaign (numeric)

poutcome: Outcome of the previous campaign (categorical)

Output Variable:
y: Indicates if the client subscribed to a term deposit (binary: "yes", "no")

This business understanding phase lays the groundwork for developing a robust predictive solution, enabling the bank to strategically prioritize leads, streamline marketing resources, and boost term deposit conversions.


## Hypothesis

🔹 Hypothesis (H₁) – Alternative Hypothesis:
There is a statistically significant relationship between the features of the client (such as age, job type, education, loan status, balance, and campaign interaction data) and the likelihood of subscribing to a term deposit. Specifically, these features can be used to predict with meaningful accuracy whether a customer will say "yes" to a term deposit offer.

🔹 Null Hypothesis (H₀):
There is no statistically significant relationship between the client attributes or campaign-related variables and the customer's decision to subscribe to a term deposit. That is, any observed patterns are due to random chance and cannot be reliably used for prediction.

## Data Preparation

Split data into input (x) and target (y) features

```python
# Drop unnecessary columns and split the data
X = dfbankadditionalfull.drop(['y'], axis=1)  # Features
y = dfbankadditionalfull['y']  # Target variable

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2,  stratify=y, random_state=42)
```

```python
# Print the shapes of the train and validation sets
print("X_train shape:", X_train.shape)
print("y_train shape:", y_train.shape)
print("X_test shape:", X_test.shape)
print("y_test shape:", y_test.shape)
```

### Label Encoder

```python
# Fit and transform the label encoder on y_train
label_encoder = LabelEncoder()
y_train_encoded = label_encoder.fit_transform(y_train)
y_test_encoded = label_encoder.transform(y_test)
```

```python
# Use sklearn's selector for safety
from sklearn.compose import ColumnTransformer, make_column_selector as selector
X_num_cols = selector(dtype_include=np.number)(X_train)
X_cat_cols = selector(dtype_include=object)(X_train)
```

### Feature Engineering

```python
input_features = X.columns
input_features
```

```python
# LogTransformer class
class LogTransformer:
    def __init__(self, constant=1):
        self.constant = constant
 
    def transform(self, X_train):
        return np.log1p(X_train + self.constant)
 
 
# Numerical transformer with LogTransformer
numerical_pipeline = Pipeline(steps=[
    ('num_imputer', SimpleImputer(strategy='median')),
    ('log_transform', FunctionTransformer(LogTransformer().transform)),
    ('scaler', StandardScaler())
])
 
class BooleanToStringTransformer(TransformerMixin):
    def fit(self, X, y=None):
        # Fit logic here, if needed
        return self
 
    def transform(self, X):
        # Transformation logic here
        # Ensure to return the transformed data
        return X.astype(str)
 
 
# Categorical transformer
categorical_pipeline = Pipeline(steps=[
    ('bool_to_str', BooleanToStringTransformer()),
    ('cat_imputer', SimpleImputer(strategy='most_frequent')),
    ('cat_encoder', OneHotEncoder())
])
 
 
# Combine transformers
preprocessor = ColumnTransformer(
    transformers=[
        ('num', numerical_pipeline, X_num_cols),
        ('cat', categorical_pipeline, X_cat_cols)
    ])
```

```python


X_train_preprocessed = preprocessor.fit_transform(X_train)
X_test_preprocessed = preprocessor.transform(X_test)

# Step 2: Convert to dense (if sparse) and check for NaNs
import scipy

if scipy.sparse.issparse(X_train_preprocessed):
    X_train_preprocessed = X_train_preprocessed.toarray()
    X_test_preprocessed = X_test_preprocessed.toarray()

# Ensure no NaNs (extra safety)
X_train_preprocessed = pd.DataFrame(X_train_preprocessed).fillna(0)
X_test_preprocessed = pd.DataFrame(X_test_preprocessed).fillna(0)

# Step 3: Feature selection
selector = SelectKBest(score_func=mutual_info_classif, k='all')
X_train_selected = selector.fit_transform(X_train_preprocessed, y_train_encoded)
X_test_selected = selector.transform(X_test_preprocessed)
```

## Modelling and Evaluation

```python
# List of models to evaluate
models = [
    ('Decision Tree', DecisionTreeClassifier(random_state=42)),
    ('Logistic Regression', LogisticRegression(max_iter=1000, random_state=42)),
    ('KNN', KNeighborsClassifier()),
    ('SVM', SVC(probability=True, random_state=42)),
    ('SGD', SGDClassifier(random_state=42)),
    ('Random Forest', RandomForestClassifier(random_state=42)),
    ('Gradient Boosting', GradientBoostingClassifier(random_state=42))  # use classifier here
]
```

#### Train and Compare Models - Unbalanced

```python
# Run and evaluate
results = []

for name, clf in models:
    print(f"\nTraining {name}...")
    clf.fit(X_train_selected, y_train_encoded)
    y_pred = clf.predict(X_test_selected)

    report = classification_report(y_test_encoded, y_pred, output_dict=True)
    results.append({
        'Model': name,
        'Accuracy': report['accuracy'],
        'Precision': report['weighted avg']['precision'],
        'Recall': report['weighted avg']['recall'],
        'F1': report['weighted avg']['f1-score']
    })

# View results
results_df = pd.DataFrame(results).sort_values(by='F1', ascending=False)
print(results_df)
```

**Insight📣**

### Visualisualizing ROC - Overlapping

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.metrics import classification_report, roc_auc_score, roc_curve
from imblearn.pipeline import Pipeline as imPipeline
from imblearn.over_sampling import SMOTE
from sklearn.feature_selection import SelectKBest, mutual_info_classif

# Create an empty results table
balanced_data_results_table = pd.DataFrame(columns=["Model Name", "Precision", "Recall", "Accuracy", "F1_Score", "AUC"])
balanced_pipelines = {}
roc_curve_data = {}

# Loop through models
for model_name, classifier in models:
    pipeline = imPipeline(steps=[
        ('preprocessor', preprocessor),
        ('smote', SMOTE(random_state=42, sampling_strategy="auto")),
        ('feature_selection', SelectKBest(mutual_info_classif, k=15)),
        ('classifier', classifier)                
    ])

    # Train the model
    pipeline.fit(X_train, y_train)

    # Predict class labels
    y_pred = pipeline.predict(X_test)

    # Map 'no' -> 0, 'yes' -> 1
    label_map = {'no': 0, 'yes': 1}
    y_pred = pd.Series(y_pred).map(label_map).values
    y_test_mapped = pd.Series(y_test).map(label_map).values
    y_pred = y_pred.astype(int)

    # Try to get probabilities or decision function scores
    try:
        y_prob = pipeline.predict_proba(X_test)[:, 1]
        auc = roc_auc_score(y_test_mapped, y_prob)
        fpr, tpr, threshold = roc_curve(y_test_mapped, y_prob)
    except AttributeError:
        try:
            y_score = pipeline.decision_function(X_test)
            auc = roc_auc_score(y_test_mapped, y_score)
            fpr, tpr, threshold = roc_curve(y_test_mapped, y_score)
        except AttributeError:
            auc = None
            fpr, tpr, threshold = None, None, None

    # Classification metrics
    balanced_metrics = classification_report(y_test_mapped, y_pred, output_dict=True)

    accuracy = balanced_metrics['accuracy']
    precision = balanced_metrics['weighted avg']['precision']
    recall = balanced_metrics['weighted avg']['recall']
    f1_score = balanced_metrics['weighted avg']['f1-score']

    # Store pipeline
    balanced_pipelines[model_name] = pipeline

    # Add results row
    balanced_data_results_table.loc[len(balanced_data_results_table)] = [
        model_name, precision, recall, accuracy, f1_score, auc
    ]

    # Store ROC data if available
    if fpr is not None and tpr is not None:
        roc_curve_df = pd.DataFrame({'False Positive Rate': fpr, 'True Positive Rate': tpr, 'Threshold': threshold})
        roc_curve_data[model_name] = roc_curve_df

# Sort results by F1 Score
balanced_data_results_table = balanced_data_results_table.sort_values(by='F1_Score', ascending=False)

# Display results table
print(balanced_data_results_table)

# Plot ROC curves
plt.figure(figsize=(8, 6))
for model_name in roc_curve_data:
    roc_data = roc_curve_data[model_name]
    auc_value = balanced_data_results_table[balanced_data_results_table["Model Name"] == model_name]["AUC"].values[0]
    plt.plot(roc_data['False Positive Rate'], roc_data['True Positive Rate'], label=f'{model_name} (AUC = {auc_value:.3f})')

# Random guess line
plt.plot([0, 1], [0, 1], color='red', lw=2, linestyle='--')

plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic (ROC) Curve')
plt.legend(loc='lower right')
plt.grid(True)
plt.tight_layout()
plt.show()
```

**Insight📣**

- The Logistic Classifier emerges as the top performer from the depicted AUC curve analysis, boasting an impressive AUC of 0.911. Its robust discriminatory power underscores its effectiveness in accurately distinguishing between churned and non-churned customers across various threshold levels, making it a reliable choice for churn prediction.

- Although slightly trailing behind the Logistic Classifier, the Random Forest Classifier still demonstrates strong performance with an AUC of 0.906. This signifies its proficiency in classifying churned and non-churned customers, albeit with a marginally lower discriminatory power compared to the Logistic Classifier.

In conclusion, opting for the Logistic Classifier as the primary model for predicting customer churn offers the dual benefit of effectively identifying potential churners while minimizing false positives. This strategic approach maximizes predictive accuracy and enhances decision-making in customer retention efforts.

### Hyperparameter Tuning

```python
import pandas as pd
import numpy as np
import joblib
import time

from sklearn.model_selection import RandomizedSearchCV
from sklearn.metrics import classification_report

# 🔹 Ensure X is a DataFrame and y is a Series with aligned indexes
X_train = X_train.reset_index(drop=True)
y_train_encoded = pd.Series(y_train_encoded).reset_index(drop=True)

# 🔹 Sample 20% of training data
sampled_indices = X_train.sample(frac=0.2, random_state=42).index
X_train = X_train.loc[sampled_indices]
y_train_encoded = y_train_encoded.loc[sampled_indices]


# 🔹 Reduced and faster parameter grids
param_grids = {
    'Decision Tree': {
        'classifier__max_depth': [None, 10],
        'classifier__min_samples_split': [2],
        'classifier__min_samples_leaf': [1]
    },
    'Logistic Regression': {
        'classifier__C': [1],
        'classifier__penalty': ['l2'],
        'classifier__solver': ['liblinear']
    },
    'KNN': {
        'classifier__n_neighbors': [3, 5],
        'classifier__weights': ['uniform']
    },
    'SVM': {
        'classifier__C': [1],
        'classifier__kernel': ['rbf']
    },
    'SGD': {
        'classifier__alpha': [0.001],
        'classifier__penalty': ['l2'],
        'classifier__loss': ['hinge'],
        'classifier__max_iter': [100]
    },
    'Random Forest': {
        'classifier__n_estimators': [100],
        'classifier__max_depth': [None],
        'classifier__min_samples_split': [2]
    },
    'Gradient Boosting': {
        'classifier__n_estimators': [100],
        'classifier__learning_rate': [0.1],
        'classifier__max_depth': [3]
    }
}

# 🔹 Initialize results DataFrame
tuned_models_df = pd.DataFrame(columns=["Model Name", "Precision", "Recall", "Accuracy", "F1_Score"])

# 🔹 Main tuning loop
for model_name, pipeline in balanced_pipelines.items():
    print(f"\n🔧 Tuning {model_name}...")
    param_grid = param_grids.get(model_name, {})
    start = time.time()

    search = RandomizedSearchCV(
        estimator=pipeline,
        param_distributions=param_grid,
        n_iter=5,             # Fewer parameter combos
        scoring='f1_weighted',
        cv=2,                 # Lower CV fold count
        random_state=42,
        n_jobs=-1,
        verbose=1
    )

    # 🔹 Train with randomized search
    search.fit(X_train, y_train_encoded)
    best_pipeline = search.best_estimator_
    balanced_pipelines[model_name] = best_pipeline

    # 🔹 Save trained pipeline
    joblib.dump(best_pipeline, f"{model_name}_pipeline.pkl")

    # 🔹 Predict on test set
    y_pred = best_pipeline.predict(X_test)

    # 🔹 Optional: remap 'yes'/'no' to 1/0
    label_map = {'no': 0, 'yes': 1}
    if isinstance(y_pred[0], str):
        y_pred = pd.Series(y_pred).map(label_map).values
    y_pred = y_pred.astype(int)

    y_test_mapped = pd.Series(y_test).map(label_map).values.astype(int)

    # 🔹 Evaluate
    metrics = classification_report(y_test_mapped, y_pred, output_dict=True)
    accuracy = metrics['accuracy']
    precision = metrics['weighted avg']['precision']
    recall = metrics['weighted avg']['recall']
    f1_score = metrics['weighted avg']['f1-score']

    # 🔹 Store results
    tuned_models_df.loc[len(tuned_models_df)] = [model_name, precision, recall, accuracy, f1_score]

    end = time.time()
    print(f"✅ {model_name} tuning took {end - start:.2f} seconds.")

# 🔹 Sort and show results
tuned_models_df = tuned_models_df.sort_values(by='F1_Score', ascending=False)
print("\n📊 All model tuning complete. Sorted results:")
print(tuned_models_df)

# 🔹 Save results to CSV
tuned_models_df.to_csv("all_model_tuning_results.csv", index=False)
```
